```python
import turtle as Pen

Pen.title ("Plan view of BZ of a Square Lattice")       # Set the graphics page Title

#Drawing the axes and the origin 

Pen.penup()                                             # Moves without writing
Pen.goto(-30,0)
Pen.pendown()                                           # Moves and writes
Pen.write("Γ",font=("Arial", 20, "normal"))             # Writing text with font type and size 
Pen.penup()
Pen.goto(0,0)

Pen.pendown()
Pen.pensize(5)                                          # Size of the drawing/writing
Pen.pencolor('black')                                   # Setting text or drawing color
Pen.goto(280,0)
Pen.stamp()                                             # Stamp arrow in front of the axes
Pen.write("Kx",font=("Arial", 16, "normal"))

Pen.penup()
Pen.goto(0,0)
Pen.pendown()
Pen.pensize(5)
Pen.pencolor('black')
Pen.goto(0,280)
Pen.left(90)                                            # rotational motion of the pen, angle 90 degree
Pen.stamp() 
Pen.right(90)
Pen.write("Ky",font=("Arial", 16, "normal"))

#Drawing the Plan view
Pen.penup()
Pen.goto(-200,-200)
Pen.pendown()
Pen.pensize(5)
Pen.pencolor('cyan')
Pen.forward(400)                                        # translational motion of the pen
Pen.left(90)
Pen.forward(400)
Pen.left(90)
Pen.write("M",font=("Arial", 16, "normal"))
Pen.forward(400)
Pen.left(90)
Pen.forward(400)
Pen.left(90)
Pen.penup()
Pen.goto(0,0)
Pen.pendown()
Pen.pensize(5)
Pen.pencolor('yellow')
Pen.goto(200,200)

# Setting values
Pen.penup()
Pen.goto(-300,0)
Pen.pendown()
Pen.pencolor('black')
Pen.write("-Π/a",font=("Arial", 20, "normal"))
Pen.penup()
Pen.goto(220,0)
Pen.pendown()
Pen.write("Π/a",font=("Arial", 20, "normal"))
Pen.penup()
Pen.goto(20,200)
Pen.pendown()
Pen.write("Π/a",font=("Arial", 20, "normal"))

Pen.bgcolor(0.5, 0.0, 1.0)                              # Background Color

Pen.penup()
Pen.goto(-400,-300)
Pen.pendown()
Pen.write("Figure: A Plan view of the Brilloin Zone of a Square Lattice",font=("Arial", 20, "bold"))    # Caption
```
