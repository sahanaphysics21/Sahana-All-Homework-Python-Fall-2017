```python
# Deltoid Curve

import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 62800)   
# 62800 gives the no. of points; (initial, final, no of points in the interval)

X = np.zeros(shape=(62800,1), dtype=float)
Y = np.zeros(shape=(62800,1), dtype=float)
# initializing two arrays of "62800" elements, each element has zero value

# print(X,Y); shows the arrays X and Y 

# for loop because we take the i-th element of theta, calculate the cosine for the 
# i-th element and store it in the corresponding ith element of array X and Y

for i in range(62800):
    X[i] = 2*np.cos(theta[i])+np.cos(2*theta[i])
    Y[i] = 2*np.sin(theta[i])-np.sin(2*theta[i])

plt.plot(X, Y, 'c')     #plot
plt.show()                       #show plot
```

```python
# Gallilean Spiral

import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 10*np.pi, 31400)   
# 31400 gives the no. of points; (initial, final, no of points in the interval)

X = np.zeros(shape=(31400,1), dtype=float)
Y = np.zeros(shape=(31400,1), dtype=float)
r = np.zeros(shape=(31400,1), dtype=float)

# initializing three arrays of "31400" elements, each element has zero value

# for loop because we take the i-th element of theta, calculate the cosine for the 
# i-th element and store it in the corresponding ith element of array X, Y and r

for i in range(31400):
    r[i] = theta[i]**2
    X[i] = r[i]*np.cos(theta[i])
    Y[i] = r[i]*np.sin(theta[i])
    

plt.plot(X, Y, 'magenta')     #plot
plt.show()                       #show plot
```

```python
# Fey's function

import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 10*np.pi, 31400)   
# 31400 gives the no. of points; (initial, final, no of points in the interval)

X = np.zeros(shape=(31400,1), dtype=float)
Y = np.zeros(shape=(31400,1), dtype=float)
r = np.zeros(shape=(31400,1), dtype=float)

# initializing two arrays of "20" elements, each element has zero value

# for loop because we take the i-th element of deg, calculate the cosine for the 
# i-th element and store it in the corresponding ith element of array X and Y

for i in range(31400):
    r[i] = np.exp(np.cos(theta[i]))-2*np.cos(4*theta[i])+(np.sin(theta[i]/12))**5
    X[i] = r[i]*np.cos(theta[i])
    Y[i] = r[i]*np.sin(theta[i])
    

plt.plot(X, Y, 'deeppink')     #plot
plt.show()                       #show plot
```
