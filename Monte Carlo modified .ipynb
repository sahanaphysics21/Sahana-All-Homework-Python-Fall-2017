```python
import numpy as np
import random as rnd
import pylab as py

def up(y):
    return y+1

def down(y):
    return y-1

def left(x):
    return x-1

def right(x):
    return x+1

def interval(x,y):
    
    step = rnd.random()
    #print(step)
    if(step > 0 and step <= 0.25):
        if(y<L):
            #print('going up')
            y = up(y)

        if(y >= L):
            interval(x,y)
    
    if(step > 0.25 and step <= 0.5):
            if(y>0):
                #print('going down')
                y = down(y)

            if(y <= 0):
                interval(x,y)
        
    
    if(step>0.5 and step <= 0.75):
            if(x>0):
                   #print('going left')
                x = left(x)

            if(x <= 0):
                interval(x,y)
    
    if(step>0.75):
        if(x < L):
            #print('going right')
            x = right(x)

        if(x >= L):
            interval(x,y)
    
    return x,y
    
    


L = 101 # Lenght of the square grid in each direction

N = int(input('Enter the number of steps ')) # number of steps

times = np.arange(0,N,1) # array of times

# x = np.zeros(N, int) # value of x
# y = np.zeros(N, int) # value of y

# initial positon
x = int(L/2)
y = int(L/2)

# trajectory of the points
xpoints = []
ypoints=[]


for t in times:
    x,y = interval(x,y)
    xpoints.append(x)
    ypoints.append(y)
    
    
# Plot the trajectory    
py.plot(xpoints,ypoints)
py.xlim(0, L)
py.ylim(0, L)
py.xlabel("Left-Right")
py.ylabel("Up-Down")
py.show()
```
