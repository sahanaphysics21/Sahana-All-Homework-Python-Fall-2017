# 5. Empty Lattice Approximation of the plan view shown above (In 5.)
### Energy Band diagram of Simple Cubic lattice 2D (square Lattice)
### Lattice type: Square lattice 


```python
# E-K diagram of Simple Cubic Lattice from plan view

import numpy as np
import matplotlib.pyplot as plt

def SCLattice(a, m, n): # a is the lattice spacing of Simple Cubic Lattice; m,n quantum numbers

    hbar = 6.582119514e-16    #Units in electron volt second (eV.s)
    me = 9.1e-31              #mass of electron in kilograms
    A = ((hbar**2))/(2 *(me))
    kx1 = ky1 =np.linspace(0, np.pi/a, 100)
    kx3 = ky2 = np.linspace(np.pi/a, 0, 100)
    ky3 = 0
    kx2 = np.pi/a
    
    G = 2*np.pi/a             #Reciprocal Lattice Vector
    
    e1 = np.empty(shape=(100, 1))                #Initialization of Energy within gamma and M
    e2 = np.empty(shape=(100, 1))                #Initialization of Energy within M and X
    e3 = np.empty(shape=(100, 1))                #Initialization of Energy within X and gamma
    
    for i in range(100):                                     #Calculating Energy for given wavenumber and quantum states
        e1[i] = A*((kx1[i]- n*G)**2 + (ky1[i] - m*G)**2)
        e2[i] = A*((kx2 - n*G)**2 + (ky2[i] - m*G)**2)
        e3[i] = A*((kx3[i]- n*G)**2 + (ky3 - m*G)**2)
        
    # print(e1)                                 # For checking the final values of Energy
    # print(e2)
    # print(e3)
    
    E1 = np.concatenate([e1, e2, e3])           # Combination of energy values in three region
    scale = a**2/(np.pi)**2                     # Rescaling in terms of (hbar**2)(np.pi**2)/(2*me)*(a**2)
    E = (E1*scale)/A
    
    #print(E)

    return E


E1 = SCLattice(3.7, 0.0, 0.0)                     
E2 = SCLattice(3.7, 1.0, 0.0)
E3 = SCLattice(3.7, 0.0, 1.0)
E4 = SCLattice(3.7, 1.0, 1.0)

X = np.linspace(0, 3, 300)                   # Point 0 is gamma, Point 1 is M, Point 2 is X, Point 3 is gamma

plt.plot(X, E1, "c")
plt.plot(X, E2, "red")
plt.plot(X, E3, "blue")
plt.plot(X, E4, "darkmagenta")
plt.xlabel("Wave Vector K") 
plt.ylabel("Rescaled Energy Levels")
plt.title("Energy Band Diagram according to Empty Lattice Approximation of Square Lattice")   
plt.show()
```
